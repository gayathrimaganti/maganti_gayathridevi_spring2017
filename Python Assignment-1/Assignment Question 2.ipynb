```python
print("The content from file 'melville-moby_dick.txt' is as below:")
string=nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
print(string)
```

```python
moby=nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
```

```python
lowerCase=moby.lower()
lowerCase
```

```python
import collections
letters = collections.defaultdict(int)
for letter in lowerCase:
    letters[letter] += 1
letters
```

```python
wordSort=sorted(letters.items(), key=lambda x: x[1], reverse=True)
print("Letters with highest frequencies are: ")
wordSort[:100]
```
