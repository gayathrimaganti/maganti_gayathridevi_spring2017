# 100 Most Common words used in the body of the email

````python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

with open("body.txt", "r") as f:
    data = f.read()
words = word_tokenize(data)
words2 = re.findall(r'(\b[a-z]{6,12}\b)',data )
common_words = [word for word in words2 if word not in stopwords.words('English')]
freq = nltk.FreqDist(common_words)
print(freq.most_common(40))
````
